```python
import pandas as pd, glob, os, time

RUTA = "/content/drive/MyDrive/monitor-fletes/datos"
archivos = sorted(glob.glob(f"{RUTA}/EstadisticasRNDC_*.xlsx"))
print(f"{len(archivos)} archivos encontrados\n")

esquemas, filas = {}, {}
t0 = time.time()
for ruta in archivos:
    mes = os.path.basename(ruta).replace("EstadisticasRNDC_", "").replace(".xlsx", "")
    df = pd.read_excel(ruta)
    esquemas[mes] = list(df.columns)
    filas[mes] = len(df)
    print(f"{mes}: {len(df):>7} filas, {len(df.columns)} columnas")
print(f"\nLeer los {len(archivos)} archivos tardó {time.time()-t0:.0f} segundos")
```

```python
# ¿Todos los meses comparten EXACTAMENTE las mismas columnas?
base = set(esquemas[list(esquemas)[0]])
print("=== Diferencias de esquema respecto al primer mes ===")
iguales = True
for mes, cols in esquemas.items():
    faltan, sobran = base - set(cols), set(cols) - base
    if faltan or sobran:
        iguales = False
        print(f"{mes}:  faltan {faltan or '—'}   sobran {sobran or '—'}")
if iguales:
    print("Idéntico conjunto de columnas en los", len(esquemas), "meses.")

# ¿El conteo de filas es estable o hay meses sospechosos?
s = pd.Series(filas)
print(f"\nFilas — mín {s.min():,} ({s.idxmin()})  máx {s.max():,} ({s.idxmax()})  mediana {int(s.median()):,}")
```

```python
import pandas as pd, glob, os
from pathlib import Path

RUTA = "/content/drive/MyDrive/monitor-fletes/datos"
SALIDA = Path("/content/drive/MyDrive/monitor-fletes/procesado")
SALIDA.mkdir(exist_ok=True)

# columnas que el índice de fletes realmente usa (de las 29, nos quedamos con estas)
COLS = ["MES", "CONFIG_VEHICULO", "CODMUNICIPIOORIGEN", "MUNICIPIOORIGEN",
        "CODMUNICIPIODESTINO", "MUNICIPIODESTINO", "CODMERCANCIA", "MERCANCIA",
        "VIAJESTOTALES", "KILOGRAMOS", "KILOMETROS", "VALORESPAGADOS",
        "VIAJESVALORCERO", "VIAJESLIQUIDOS", "GALONES",
        "KILOMETROSREGRESO", "KILOGRAMOSREGRESO"]

for ruta in sorted(glob.glob(f"{RUTA}/EstadisticasRNDC_*.xlsx")):
    mes = os.path.basename(ruta).replace("EstadisticasRNDC_", "").replace(".xlsx", "")
    destino = SALIDA / f"fletes_{mes}.parquet"
    if destino.exists():
        print(f"{mes}: ya existe, salto")          # no re-procesa lo ya hecho
        continue
    df = pd.read_excel(ruta)[COLS]
    df.to_parquet(destino, index=False)
    print(f"{mes}: {len(df):,} filas → {destino.name} ({destino.stat().st_size/1e6:.1f} MB)")

print("\nListo. A partir de ahora se lee el Parquet, no el Excel.")
```

```python
import pandas as pd, glob

# Lee TODOS los meses de una (ahora es instantáneo)
df = pd.concat([pd.read_parquet(f) for f in
                sorted(glob.glob("/content/drive/MyDrive/monitor-fletes/procesado/*.parquet"))],
               ignore_index=True)
df["periodo"] = pd.to_datetime(df["MES"].astype(str), format="%Y%m")
print(f"{len(df):,} filas, {df['periodo'].nunique()} meses "
      f"({df['periodo'].min():%Y-%m} a {df['periodo'].max():%Y-%m})")

# Filtros del contrato v1 (los que calibraste a golpes en la exploración)
pes = df[df["CONFIG_VEHICULO"].str.contains("Tractocam", case=False, na=False) &
         (df["VALORESPAGADOS"] > 0) & (df["VIAJESVALORCERO"] == 0) &
         (df["VIAJESLIQUIDOS"] == 0) & (df["KILOMETROS"] >= 200) &
         (df["CODMUNICIPIOORIGEN"] != df["CODMUNICIPIODESTINO"])].copy()

def flete_mensual(g):
    return g["VALORESPAGADOS"].sum() / (g["KILOGRAMOS"].sum()/1000 * g["KILOMETROS"].mean())

# Serie mensual de $/ton-km para Buenaventura→Bogotá
pes["corredor"] = pes["MUNICIPIOORIGEN"] + " → " + pes["MUNICIPIODESTINO"]
bunbog = pes[pes["corredor"].str.contains("BUENAVENTURA.*BOGOT", case=False, na=False)]
serie = bunbog.groupby("periodo").apply(flete_mensual, include_groups=False).rename("flete_ton_km")
print(serie.round(0))
serie.plot(figsize=(11,4), marker="o",
           title="Flete $/ton-km — Buenaventura→Bogotá (RNDC, pesados)")
```
